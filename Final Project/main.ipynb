### Figure 2. Anxiety by Duration Group (T-test)

Following the linear correlations, listening duration was dichotomized to test for potential group-level differences in anxiety. This analysis tests the hypothesis that individuals with longer daily music listening duration exhibit lower anxiety levels than those with shorter listening duration. Although the T-test indicates a statistically significant difference (p = 0.030), the distributions overlap substantially, indicating a small effect size with limited practical relevance. This suggests that duration alone is unlikely to be a meaningful driver of anxiety differences.

### Figure 3. Interaction Effects of Duration with BPM and Age

Given the limited main effects, interaction analyses were conducted to assess whether BPM or age moderates the relationship between duration and anxiety. While slight trends are visible, wide and overlapping confidence intervals indicate no robust interaction effects. These results further suggest that simple two-way combinations of variables do not explain anxiety well.
