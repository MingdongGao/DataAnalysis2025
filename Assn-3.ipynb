### Figure 2

Use `data-fig2.npy` for this figure.

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.load( 'data-fig2.npy' )

x = np.arange(a.shape[0])
y_red   = a[:, 0]
y_green = a[:, 1]
y_blue  = a[:, 2]

plt.figure()
plt.plot(x, y_red,   color='red',   label='Red')
plt.plot(x, y_green, color='green', label='Green')
plt.plot(x, y_blue,  color='blue',  label='Blue')

plt.xlabel('X value')
plt.ylabel('Y value')
plt.legend()
plt.show()
```

### Figure 3

Use `data-fig3.npy` for this figure.

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.load( 'data-fig3.npy' )

x = a[:, 0]
y = a[:, 1]

plt.figure()
plt.scatter(x, y, s=300, color='blue', alpha=0.3)  

plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

### Figure 4

Use `data-fig4.npy` for this figure.

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.load( 'data-fig4.npy' )

categories = ['A', 'B', 'C', 'D', 'E']

plt.figure()
plt.bar(categories, a, color='gray', alpha=0.5)

plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.load( 'data-fig5.npy' )

plt.figure(figsize=(8, 6))
plt.imshow(a, cmap='RdBu', origin='lower',
           aspect='auto', vmin=-1, vmax=1)
plt.colorbar() 

plt.show()
```
